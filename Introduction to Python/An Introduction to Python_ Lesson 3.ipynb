### **-> if, elif and else statements**

### **-> for and while loops**

```
prob_rain = 0.2
if prob_rain >= 0.3:
  print("carry an umbrella")
```

```
name ="zappy"

if name is "dawn":
  print("welcome dawn")
elif name is "shawn":
  print("what up shawny")
else:
  print("hello stranger")
```

```
foods = ["cheese", 'oranges', 'nutella', 'coke']

for f in foods:
  print(f)
```

```
for x in range (5):
  print (x)
```

```
t = 4

while t < 10:
  print(t)
  t = t + 1
```
