## **Variables and Strings**

**Variable:**  A Python variable is a reserved memory location to store values. In other words, a variable in a python program gives data to the computer for processing.

**String:** Strings are arrays of bytes representing Unicode characters. Python does not have a character data type, and a single character is simply a string with a length of 1.
